## Data Import

```python
print("Loading data...")
data_path = ".\data\dataset"
original_x_train, original_x_test, original_y_train, train_ids, test_ids = load_csv_data(data_path)
```

## Train Test split

```python
from evaluation import train_test_split
X_train, X_test, y_train, y_test = train_test_split(original_x_train, original_y_train)
```

## First model

- Preprocessing : Missing values imputation using the mean value, + scaling
- Data balancing : Generation of nine balanced subsets of the training data set using duplication.
- Learning Algorithm : The final prediction is computed by averaging the different probabilities output by nine logistic regression models trained on each one of the previous subsets

We perform a grid search, to look for the optimal degree for the polynomial feature extension as well as the optimal threshold for decision-making.

```python
thresholds = np.linspace(0.5, 0.8, 60)
plot_f1_accuracy_3d(X_train, X_test, y_train, y_test, thresholds)
```

## Second model

- Preprocessing : Missing values imputation using the mean value, + scaling
- Data balancing : Generation of seven balanced subsets of the training data set using synthetization.
- Learning Algorithm : The final prediction is computed by averaging the different probabilities output by nine logistic regression models trained on each one of the previous subsets

We do the same grid search as before

```python
thresholds = np.linspace(0.5, 0.8, 60)
plot_f1_accuracy_3d(X_train, X_test, y_train, y_test, thresholds, synt_data_method=True)
```

## Third model

- Preprocessing : Missing values imputation using the mean value for numerical columns. Regarding categorical columns, the imputation was done using the mode. Afterwards, levels are bring to their minimal possible values. Finally, frequency encoding is used to fit the values within $[0, 1]$.
- Data balancing : Generation of nine balanced subsets of the training data set using duplication.
- Learning Algorithm : The final prediction is computed by averaging the different probabilities output by nine logistic regression models trained on each one of the previous subsets

We do the same grid search as before

```python
thresholds = np.linspace(0.5, 0.8, 60)
plot_f1_accuracy_3d(X_train, X_test, y_train, y_test, thresholds, cat_encoding=True)
```

## Conclusion

Finally, the First model has better performances. We select this model for the submission.

```python
# Submission

degree = 3
threshold = 0.68305

x_train_prepared, mean, scaling_mean, scaling_std, degree, rm_nan_columns_idx = prepare_train_data(original_x_train, degree=degree)
x_test_prepared = prepare_test_data(original_x_test, mean, scaling_mean, scaling_std, degree, rm_nan_columns_idx)

models = train_models(x_train_prepared, original_y_train)

y_pred = predict_with_voting(models, x_test_prepared, threshold)

create_csv_submission(test_ids, y_pred, name="submission_test.csv")
```
